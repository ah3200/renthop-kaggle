#### Remove outliter (unusually overpriced rental listing)

```python
train = train[train.price < 100000]
X_train = train[['bathrooms','bedrooms','price']]
#y_train = train['interest_level']
```

```python
train.shape, X_train.shape, y_train.shape
```

#### Label encoding target variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(train['interest_level'])
```

#### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
```

#### Cross validation

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(logit, X_train, y_train, cv=10, scoring='log_loss')
```
